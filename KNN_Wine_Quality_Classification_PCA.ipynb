```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
```

```python
wine_data = pd.read_csv('winequality-red.csv')
wine_data.info()
```

```python
Target = wine_data['quality']
Features = wine_data.iloc[:,0:11]
Features
```

```python
stdscaler = StandardScaler()
#mmscaler = MinMaxScaler()
#robscaler = RobustScaler()
Features = pd.DataFrame(stdscaler.fit_transform(Features),columns=Features.columns)
```

```python
Feature_Set1 = ['total sulfur dioxide', 'free sulfur dioxide', 'alcohol']
Feature_Set2 = ['alcohol', 'volatile acidity', 'total sulfur dioxide', 'sulphates']
Feature_Set3 = ['fixed acidity', 'volatile acidity']
Feature_Set4 = ['alcohol', 'sulphates', 'total sulfur dioxide', 'volatile acidity']
Feature_Set5 = ['alcohol','sulphates','total sulfur dioxide','volatile acidity','residual sugar']
Feature_Set6 = ['alcohol', 'fixed acidity', 'free sulfur dioxide']
Feature_Set7 = ['fixed acidity', 'volatile acidity', 'citric acid', 'sulphates', 'alcohol']
Feature_Set_list = [Feature_Set1,Feature_Set2,Feature_Set3,Feature_Set4,Feature_Set5,Feature_Set6,Feature_Set7,Feature_Set8,Feature_Set9]
Feature_Set_list
```

```python
pca = PCA(n_components=2)
Features_PCA = pca.fit_transform(Features)
PCA_Components=pd.DataFrame(data=Features_PCA,columns = ['Principal Components1','Principal Components2'])
```

```python
Training_Score_results=[]
Test_Score_results=[]
Best_Parameter_results=[]
for i in Feature_Set_list:
    knn = KNeighborsClassifier(n_neighbors=3)
    X_train,X_test,y_train,y_test=train_test_split(Features[i],Target,test_size=0.3,random_state=24,stratify=Target)
    knn.fit(X_train,y_train)
    Training_Score = knn.score(X_train,y_train)
    Training_Score_results.append(Training_Score)
    Test_Score = knn.score(X_test,y_test)
    Test_Score_results.append(Test_Score)
```

```python
Training_Score_results = pd.DataFrame(Training_Score_results,columns=['Training_Score'])
Test_Score_results = pd.DataFrame(Test_Score_results,columns=['Test_Score'])
Best_Parameter_results = pd.DataFrame(Best_Parameter_results)
All_Results = pd.concat([Training_Score_results,Test_Score_results,Best_Parameter_results],axis=1)
All_Results['Diff'] = All_Results['Training_Score']-All_Results['Test_Score']
All_Results = pd.concat([All_Results,pd.DataFrame(Feature_Set_list)],axis=1) 
All_Results
```

### Apply PCA to Best Features

Best Features =  ['alcohol', 'sulphates', 'total sulfur dioxide', 'volatile acidity','chlorides']

```python
pca = PCA(n_components=3)
Features_PCA_Best = pca.fit_transform(Features[['alcohol', 'volatile acidity', 'total sulfur dioxide', 'sulphates']])
PCA_Components_Best=pd.DataFrame(data=Features_PCA_Best,columns = ['Principal Components1','Principal Components2','Principal Components3'])
```

```python
training_score_results=[]
test_score_results=[]

X_train,X_test,y_train,y_test=train_test_split(PCA_Components_Best,Target,random_state=24,test_size=0.3,stratify=Target)
    
knn = KNeighborsClassifier()
parameters=dict(n_neighbors=np.arange(2,20,1))
cv = GridSearchCV(knn,param_grid=parameters,cv=3)
cv.fit(X_train,y_train)

best_params = cv.best_params_

training_score=cv.score(X_train,y_train)
training_score_results.append(training_score)
    
test_score=cv.score(X_test, y_test)
test_score_results.append(test_score)

print(training_score_results,test_score_results,best_params)
```
